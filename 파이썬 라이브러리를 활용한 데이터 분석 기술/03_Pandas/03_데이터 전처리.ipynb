### inplace
- 제공되는 전처리 관련 함수들을 inplace 매개변호를 제공하는 것들이 있다.
- inplace가 없는 함수들은 원본을 수정하거나 원본을 유지한채 작업이 반영된 새로운 것을 생성해 반환한다.
- inplace를 제공할 경우 <span style = 'color:green'>True</span> 를 넣어주면 <span style = 'color:green'>원본이 변경</span>되고 <span sytle = 'color:orange'>False</span>를 넣어주면 <span sytle = 'color:orange'>원본은 유지하고 작업이 반영된 새로운 것을 생성</span>해 반환한다.

### 조건에 해당하는 데이터를 검색해온다.

```python
# 데이터 프레임도 True나 False로 구성된 요소를 넣어주면 True에 해당하는 행들만 가져올 수 있다.
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')

# 원래는 데이터프레임[] 방식으로 접근을 하면 컬럼을 선택하는 것이지만
# True, False로 구성된 것을 넣어주면 True에 해당하는 행만 가져온다.
df1[[True, False, True, False, True]]
```

```python
# 국어 점수가 90점보다 큰
df2 = df1[df1['국어']>90]
df2
```

```python
# 검색
# 국어점수가 90점보다 큰 학생 정보를 가져온다.
df2 = df1.query('국어 > 90')
```

```python
# 성별이 남자인 학생들
# 식 안에 따옴표가 없는 문자열을 컬럼이름으로 인식하고 따옴표가 있는 것은 문자열 값으로 인식한다.
df2 = df1.query('성별 == "남자"')
df2
```

```python
# and
df2 = df1.query('국어 > 80 and 수학 > 80')
df2
```

```python
# or
df2 = df1.query('국어 < 70 or 수학 < 70')
df2
```

```python
# 변수 사용
a1 = 90
df2 = df1.query('국어 > @a1')
df2
```

```python
list1 = ['철수', '영희', '민수']

# 리스트 안에 포함되어 잇는 행만 가져온다.
df2 = df1.query('이름 in @list1')
df2
```

```python
# 리스트 안에 포함되어 있지 않은 행만 가져온다.
df2 = df1.query('이름 not in @list1')
df2
```

### 데이터 프레임 병합

```python
temp_df = pd.read_csv('data/grade.csv', encoding='euc-kr')
temp_df.set_index('이름', inplace=True)
temp_df.loc['민호'] = [1, '남자', 50, 60,70, 80]


df1 = temp_df.loc[['철수', '영희']]
df2 = temp_df.loc[['민수', '수현']]
df3 = temp_df.loc[['호영', '민호']]

display(df1)
display(df2)
display(df3)
```

```python
# append: 그냥 뒤에 붙이는 것
df4 = df1.append([df2, df3])
df4
```

```python
# df2 컬럼의 순서를 변경한다.
df2 = df2[['국어', '영어', '수학', '과학', '학년', '성별']]
display(df2)

# 그리고 다시 append를 해본다면? -> 그래도 잘 된다. key값에 따른 value가 잘 연결되어 있기 때문!
# 컬럼의 이름이 동일한 것끼리 합쳐진다.
df4 = df1.append([df2, df3])
display(df4)
```

```python
# 존재하지 않는 컬럼이 있을 경우 -> 한국사 컬럼 추가 및 다른 행들 결측치 생성.
df2['한국사'] = [100,200]
display(df2)

df4 = df1.append([df2, df3])
display(df4)
```

```python
# merge 테스트를 위해 데이터 프레임 두 개를 만든다.
a1 = {
    '고객번호' : [1001, 1002, 1003],
    '주소지' : ['서울', '부산', '대전'],
    '이름' : ['홍길동', '김길동', '최길동']
}

a2 = {
    '고객번호' : [1001,1002,1003,1004],
    '배송지' : ['서울', '부산', '대전', '대구'],
    '금액' : [10000,20000, 30000, 40000]
}

df1 = pd.DataFrame(a1)
df2 = pd.DataFrame(a2)
display(df1)
display(df2)
```

```python
# merge 사용
# 컬럼명이 동일한 컬럼을 기준으로 행을 합쳐준다.(inner join)

df10 = pd.merge(df1, df2)
df10
```

```python
# 합칠 기준이 되는 컬럼을 직접 지정한다.
df20 = pd.merge(df1, df2, left_on='주소지', right_on='배송지')
df20
```

```python
# outer join: how를 생각하면 inner join을 수행한다. inner join은 기존이 되는 컬럼의 값이 양쪽 모두 있는 것만 합쳐주고 둘 중 하나라도 없는 것은 제외한다.
# 만ㅇ약 둘 중 하나라도 없는 것까지 포함하고자 한다면 outer join을 수행한다.
df30 = pd.merge(df1, df2, how='outer')
df30
```

```python
#left outer join
# 좌측의 데이터프레임은 있지만 우측에는 없는 것만 포함한다.
df100 = pd.merge(df1, df2, how='left')
df200 = pd.merge(df2, df1, how='left')

display(df100)
display(df200)
```

```python
# right outer join
# 우측의 데이터프레임은 있지만 whk측에는 없는 것만 포함한다.
df100 = pd.merge(df1, df2, how='right')
df200 = pd.merge(df2, df1, how='right')

display(df100)
display(df200)
```
