```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 그래프를 그리기 위한 기본 설정
# 폰트설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'

# 폰트설정 - 맥
# plt.rcParams['font.family'] = 'Apple Gothic'

# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

# 글자 크기
plt.rcParams['font.size'] = 14

# 폰트 설정 시 '-'기호 깨는거 방지하기
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 100행 3열 랜덤값
x = np.random.randn(100, 3)
# 2021년 1월 1일부터 100일치 날짜를 가져온다.
index = pd.date_range('2021/01/01', periods=100)

df1 = pd.DataFrame(x, index=index, columns=['A', 'B', 'C'])
df1
```

### 선그래프

```python
# 각 컬럼의 데이터가 그래프 하나가 된다.
# 범례도 자동으로 나타나며 범례에 표시되는 이름은 컬럼의 이름으로 설정된다.
df1.plot()

plt.title('선 그래프')
plt.xlabel('시간')
plt.ylabel('Data')

plt.show()
```

### 다른 그래프 그리기
- plot 함수에 `kind`라는 매개변수에 그래프의 종류를 지정할 수 있다.
- bar, pie, hist, kde, box, scatter, area

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
iris
```

```python
# Series 하나의 데이터
iris['꽃받침길이'][:20].plot(kind='bar', rot=0)
plt.title('막대 그래프')
plt.show()
```

- plot 함수의 kind에 설정하는 문자열과 동일한 이름의 함수도 제공한다.

```python
iris['꽃받침길이'][:20].plot.bar(rot=0)
plt.title('막대 그래프')
plt.show()
```

```python
# 품종을 제외한 나머지를 가지고 그래프를 그려본다.
# 하나의 row를 추출하여 row 내의 컬럼별 막대 그래프를 그리고, 그것을 하나의 그룹으로 묶어준다.
df2 = iris.drop('품종', axis=1)

df2.head(20).plot.bar(rot=0)
plt.show()
```

```python
#  품종별 컬럼의 평균을 구한다.
iris.groupby('품종').mean().plot.bar()
plt.show()
```

### 파이 그래프

```python
a1 = iris['품종'].value_counts()
# 매개변수에 들어가는 것들은 matplotlib에서 제공하는 pie plot과 동일하다.
a1.plot.pie(autopct='%.2f%%', startangle=90)
```

```python
iris.plot.hist(alpha=0.5)
plt.show()
```

### kde

```python
iris.plot.kde()
plt.show()
```

### boxplot

```python
iris.plot.box()
plt.show()
```

```python
iris.boxplot()
plt.show()
```
